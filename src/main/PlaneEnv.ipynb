## Random Envirnoment-Executions Rewards

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PlaneEnvironment import PlaneEnv
```

```python
env = PlaneEnv('../../planes/plane1.txt')
```

```python
iters = 200
rewards = []
for i in range(iters):
    env.reset()
    done = False
    while (done == False):
        action = env.action_space_sample()
        (x, reward, done, info) = env.step(action)
        #if (i%15):
        #    print(x)
        
    #env.plane.drawPlane()
    rewards.append(reward)
    #print("iter: " + str(i) + ", reward: " + str(reward))
    #print()
```

```python
plt.plot(rewards)
```

## Q-Learning Algorithm for passengers allocation

```python
import numpy as np
from QLearningAlgorithm import QLearning
```

```python
ql = QLearning('../../planes/plane1.txt')
```

```python
s=10
```

```python
(totalrewards,ia3s) = ql.launch(epsType='t2', slot=s, N=500)
```

```python
def plots(totalrewards, ia3,slot=100):
    plt.subplots(figsize=(18,5))
    plt.subplot(121)
    plt.plot(totalrewards)
    plt.subplot(122)    
    plt.subplot(122)
    plt.plot(ia3)
    plt.title("Ia3")

    N = len(totalrewards)
    running_avg = np.empty(N)
    for t in range(N):
        running_avg[t] = totalrewards[max(0, t-slot):(t+1)].mean()
    plt.subplot(121)
    plt.plot(running_avg)
    plt.title("Rewards")
```

```python
plots(totalrewards, ia3s, slot=s)
```
